```python
import os
import re
import pandas as pd

filename = r'C:\Users\sifat\Documents\DS Study\Exercise Files\Ch01\resources\env.txt'
new_filename = r'C:\Users\sifat\Documents\DS Study\Exercise Files\Ch01\resources\env_1.txt'

if(os.path.exists(filename)):
    print('Exists')
    with open(filename) as file:
        lines = file.readlines()
    with open(new_filename, 'w') as file_1:
        for line in lines:
            line = '\t'.join(re.split(r'[\t\s]+', line))
            file_1.write(line+'\n')

if(os.path.exists(new_filename)):
    print('Exists')
    df = pd.read_csv(new_filename, sep='\t', header=0)

df.drop(columns='Unnamed: 9')
# df.shape
```
